```python
#Step 1-Data collection

import pandas as pd     #Pandas is a Python library used for data analysis. After importing Pandas, we use `pd` to refer to the library throughout the code.
```

```python
dataset=pd.read_csv("Salary_Data.csv")   # The Pandas library is used to read tabular data into Python.
                                         # The `read` functions in Pandas are used to load data files.
                                         # Here, the file should be saved in CSV format.
```

```python
dataset    
```

```python
#step 2- Data processing
```

```python
independent =dataset[["YearsExperience"]]
```

```python
independent
```

```python
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
# step 3 - spilit data for taining and testing.
#sklearn is a library used in python to split data.
```

```python
from sklearn.model_selection import train_test_split   #model_selection is a class in sklearn. 
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)  
# no need to put train_size because training is automatically take 80(1-0.20) per data to train
# randomly split data. 
```

```python
x_train   
```

```python
# to train data
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()        # to install all class or function from linearRegression we are creationg one variable to put all data.
regressor.fit(x_train,y_train)
```

```python
regressor    # to confirm that data inside function linearRegression is tranfered into regressor variable. 
```

```python
weight=regressor.coef_    
```

```python
weight
```

```python
bais=regressor.intercept_
bais
```

```python
# y=9360x+26777 is now our model.
```

```python
# next check whether model predict data using test data
```

```python
y_pred=regressor.predict(x_test) #take y prediction value using x test value using y=mx+c.
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score    #here r2_score is function
r_score=r2_score(y_test,y_pred)
```

```python
r_score     # as common if more than 0.9 is accpetable model.
```

```python
import pickle      #Pickle is a function.
mymodel1="finalized_model_linear.sav"    # file is saved in sav format within variable (mymodel1)
```

```python
pickle.dump(regressor,open(mymodel1,'wb'))    #pickle.dump used to save the model from regressor to mymodel1 in binary format. wb means binary foramt.
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))   #rb means read binary
result=loaded_model.predict([[15]])
```
